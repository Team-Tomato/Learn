```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
mal_filepath="../input/malnutrition-across-the-globe/country-wise-average.csv"
mal_data =pd.read_csv(mal_filepath)
list(mal_data.columns)
```

```python
plt.figure(figsize=(10,26))
sns.barplot(x=mal_data['Wasting'],y=mal_data['Country'])
```

```python
plt.figure(figsize=(14,11))
sns.lineplot(data=mal_data['Wasting'],label="wasting")
sns.lineplot(data=mal_data['Underweight'],label="underweight")
```

```python
sns.scatterplot(x=mal_data['Income Classification'],y=mal_data['Overweight'],hue=mal_data['Stunting'])
```
